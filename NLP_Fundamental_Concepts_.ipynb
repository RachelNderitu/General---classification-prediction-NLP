**Basic Data PreProcessing**

Demonstrate basic data preprocessing steps such :
 Tokenization,
 Lowercasing,
 Removing Stopwords, punctuations, or unwanted tokens
 Stemming,
 Lemmatization,
 PoS tagging,
 NER, and
 Chunking

```python
# 1. Tokenization
import nltk
from nltk import word_tokenize
nltk.download('punkt') # download the punkt resource
text = "Natural Language Processing is a field of study that focuses on the interaction between computers and humans using natural language."
tokens = nltk.word_tokenize(text)
print(tokens)
```

```python
# 2. Lowercasing
tokens =[w for w in word_tokenize(text.lower())if w.isalpha()]
```

```python
# 3. Remove stopwords
from nltk.corpus import stopwords
nltk.download('stopwords') # download the stopwords corpus

#no_stops =[t for t in tokens if t not in stopwords.words(english)]
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
```

```python
# 4. Removing puntuations and nonalphanumeric characters using Regular expressions
import re
myString = "<I am going to show 2? or maybe 10 examples of using regular expressions!.>"
''' returns a list of every alphanumeric character that appears in the
string above, i.e. we get a list of 57 characters from I to s'''

re.findall(r"[a-zA-Z0-9.]",myString)
'''If we are interested in words, numbers included, and not just the
individual
characters we can include a + at the end of the expression, which is
special
as it matches one or more characters in the preceding regular
expression'''

re.findall(r"[a-zA-Z0-9]+",myString)
''' We can use a special shorthand symbols to make life easier'''
# so now we know that [a-zA-Z0-9] = \w in RE
re.findall(r"\w+",myString)
```

```python
# 5. Stemming
from nltk.stem import PorterStemmer
```

```python
# 6. Lemmatization
from nltk import WordNetLemmatizer
nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in
filtered_tokens]
print(lemmatized_tokens)
```

```python
# 7. Part-of-speech (POS) Tagging
nltk.download('averaged_perceptron_tagger')
pos_tags = nltk.pos_tag(lemmatized_tokens)
print(pos_tags)

```

```python
# 8. Named Entity Recognition
import nltk
nltk.download('maxent_ne_chunker')  # Download the required data package
nltk.download('words') # Download the words corpus

text = "Barack Obama was born in Hawaii."
tokens = nltk.word_tokenize(text)
pos_tags = nltk.pos_tag(tokens)
ne_tags = nltk.ne_chunk(pos_tags)
print(ne_tags)
```
